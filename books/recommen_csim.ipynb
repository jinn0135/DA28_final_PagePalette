```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('temp_books.csv')
print(df.duplicated(['title','author','publisher','rate']).sum())
df.drop_duplicates(subset=['title','author','publisher','rate'], ignore_index=True, inplace=True)
```

```python
def get_recomm_g2(df, gen, age, title, n=15):
    ga = 2*(age//10)-2 if gen=='female' else 2*(age//10)-1
    cvect = CountVectorizer(ngram_range=(1,3))
    cvect_g2 = cvect.fit_transform(df['genre_2'])
    csim_g2 = cosine_similarity(cvect_g2, cvect_g2).argsort()[:,::-1]
    
    idx = df[df['title'] == title].index.values
    sim_idx = csim_g2[idx, :n].reshape(-1)
    sim_idx = sim_idx[sim_idx != idx]
    return df.iloc[sim_idx].sort_values(df.columns[6+ga], ascending=False)

gen, age = 'male', 20
title = '슈독 : 나이키 창업자 필 나이트 자서전'
get_recomm_g2(df, gen, age, title)
```

```python
def get_recomm_GenAge(df, gen, age, n=15):
    ga = 2*(age//10)-2 if gen=='female' else 2*(age//10)-1
    print(ga)
    target = np.array([.0 for i in range(12)])
    target[ga] = 99.9
    target = target.reshape(-1,12)
    concat_df = pd.concat([df,pd.DataFrame(target, columns=df.columns[6:].values)], ignore_index=True)
    
    csim_ga = cosine_similarity(concat_df.iloc[:,6:], concat_df.iloc[:,6:]).argsort()[:,::-1]
    
    idx = concat_df.index.values[-1]
    sim_idx = csim_ga[idx, :n].reshape(-1)
    sim_idx = sim_idx[sim_idx != idx]
    return df.iloc[sim_idx].sort_values(df.columns[6+ga], ascending=False)

gen, age = 'male', 20
get_recomm_GenAge(df, gen, age)
```
