```python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('C:/Users/Playdata/playdata_wj/Project/best_books_소설시희곡.csv')
print(df.duplicated(['title','author','rate']).sum())
df.drop_duplicates(subset=['title','author','rate'], ignore_index=True, inplace=True)
print(df.duplicated(['title','author','rate']).sum())
```

```python
def get_recomm_GenAge(df, gen, age, n=15):
    ga = 2*(age//10)-2 if gen=='female' else 2*(age//10)-1

    target = np.array([.0 for i in range(12)])
    target[ga] = 99.9
    target = target.reshape(-1,12)
    concat_df = pd.concat([df,pd.DataFrame(target, columns=df.columns[6:].values)], ignore_index=True)
    
    csim_ga = cosine_similarity(concat_df.iloc[:,6:], concat_df.iloc[:,6:]).argsort()[:,::-1]
    
    idx = concat_df.index.values[-1]
    sim_idx = csim_ga[idx, :n].reshape(-1)
    sim_idx = sim_idx[sim_idx != idx]
    return df.iloc[sim_idx].sort_values(df.columns[6+ga], ascending=False)

gen, age = 'male', 20
get_recomm_GenAge(df, gen, age)
```

```python
def get_recomm_g2(df, gen, age, title, n=15):
    ga = 2*(age//10)-2 if gen=='female' else 2*(age//10)-1
    cvect = CountVectorizer(ngram_range=(1,3))
    cvect_g2 = cvect.fit_transform(df['genre_2'])
    csim_g2 = cosine_similarity(cvect_g2, cvect_g2).argsort()[:,::-1]
    
    idx = df[df['title'] == title].index.values
    sim_idx = csim_g2[idx, :n].reshape(-1)
    sim_idx = sim_idx[sim_idx != idx]
    return df.iloc[sim_idx].sort_values(df.columns[6+ga], ascending=False)

gen, age = 'male', 20
title = '슈독 : 나이키 창업자 필 나이트 자서전'
get_recomm_g2(df, gen, age, title)
```

```python
class recommBook():
    def __init__(self, gen, age, g1, g2):
        self.ga = 2*(age//10)-2 if gen=='female' else 2*(age//10)-1
        self.g1, self.g2 = g1, g2
        print('user info: {}, {}, {}, {}'.format(gen, age, g1, g2))
        self.df = self.call_df(self.g1)
        
    def call_df(sefl, g1):
        df = pd.read_csv('C:/Users/Playdata/playdata_wj/Project/best_books_{}.csv'.format(g1))
        print(df.duplicated(['title','author','rate']).sum(), end='->')
        df.drop_duplicates(subset=['title','author','rate'], ignore_index=True, inplace=True)
        print(df.duplicated(['title','author','rate']).sum())
        return df
    
    def recomm1(self, df, n=15):
#         new_df = self.recomm1_g1(df)
        new_df = self.df.copy()
        check = True
        
        try:
            idx = new_df[new_df['genre_2']==self.g2].index.values[0]
        except:
            check = False
            new_df = pd.concat([new_df, pd.DataFrame({'genre_2':[self.g2]})], ignore_index=True)
            idx = new_df[new_df['genre_2']==self.g2].index.values

        cvect = CountVectorizer(ngram_range=(1,3))
        cvect_g2 = cvect.fit_transform(new_df['genre_2'])
        csim_g2 = cosine_similarity(cvect_g2, cvect_g2).argsort()[:,::-1]
        
        sim_idx = csim_g2[idx, :n].reshape(-1)
        sim_idx = sim_idx if check else sim_idx[sim_idx != idx]
        return new_df.iloc[sim_idx].sort_values(new_df.columns[5+self.ga], ascending=False)
        
    def recomm_GenAge(self, df, n=15):
        target = np.array([.0 for i in range(12)])
        target[self.ga] = 99.9
        target = target.reshape(-1,12)
        concat_df = pd.concat([df, pd.DataFrame(target, columns=df.columns[6:].values)], ignore_index=True)

        csim_ga = cosine_similarity(concat_df.iloc[:,6:], concat_df.iloc[:,6:]).argsort()[:,::-1]

        idx = concat_df.index.values[-1]
        sim_idx = csim_ga[idx, :n].reshape(-1)
        sim_idx = sim_idx[sim_idx != idx]
        return df.iloc[sim_idx].sort_values(df.columns[5+self.ga], ascending=False)
    
    def recomm_g2(self, df, title, n=15):
        cvect = CountVectorizer(ngram_range=(1,3))
        cvect_g2 = cvect.fit_transform(df['genre_2'])
        csim_g2 = cosine_similarity(cvect_g2, cvect_g2).argsort()[:,::-1]

        idx = df[df['title'] == title].index.values
        sim_idx = csim_g2[idx, :n].reshape(-1)
        sim_idx = sim_idx[sim_idx != idx]
        return df.iloc[sim_idx].sort_values(df.columns[5+self.ga], ascending=False)
```
